### План исследования

1. Подготовка данных
   - Загрузить данные о клиентах, заказах и коммуникациях - `data specification` (понимание данных)

   - Объединение данных

 2. Исследование базы клиентов
   - Оценить общее количество клиентов и их распределение по брендам и ресторанам

   - Определить активность клиентов (как часто они делают заказы, какой вид заказов предпочитают и т.д., различаются ли профили клиентов доставки-самовывоза-alacart), есть ли клиенты, которые предпочитают только один вид заказов и не пользуются другими
   - Построить RFM-сегментацию клиентов и посмотреть профиль каждого макросегмента / группы сегментов, распределение сегментов по ресторанам

 3. Анализ заказов - `Уровень покупательской активности` (количество, частота, размер и состав покупок)
   - Распределение признаков по брендам 
   - Распределение признаков по ресторанам 
   - Изучить средний чек заказа
   - Оценить предпочтения клиентов по блюдам и напиткам
   - Выявить популярность различных способов получения заказа (в ресторане, самовывоз, доставка)

   - аналитика гостепотока: evernote:///view/230155620/s577/b7104694-981e-23b6-39da-abd1f94e1a16/77159425-b2c3-4fb2-114c-af2b8cb7fb3d

4. Анализ коммуникаций с клиентами - `Уровень коммуникационной активности` (реакция на сообщения, переходы, авторизации, участие в мероприятиях)
   - Оценить эффективность коммуникаций посредством смс, емейл, пуш уведомлений
   - Провести сегментацию клиентов и оценить, какой вид коммуникации наиболее эффективен для каждой группы клиентов



5. Прогнозирование поведения клиентов
   - Построить модель прогнозирования оттока клиентов
   - Изучить факторы, влияющие на вероятность оттока, и предложить меры по его снижению



6. Визуализация данных
   - Построить графики и диаграммы для наглядного представления результатов исследования
   - Подготовить презентацию с ключевыми выводами и рекомендациями для управления ресторанной сетью



7. Выводы и рекомендации
   - Сформулировать основные выводы и предложить рекомендации по улучшению работы с клиентами и повышению эффективности бизнеса.

8. Подготовка к построению моделей ML
- Определение бизнес-задач, которые будут решать модели
    - `LTV` - сегментация клиентов по ltv (какими еще характеристиками они обладают, чтобы с самого начала взаимодействия знать, что у этого клиента будет большое ltv, и ему можно давать входные скидки
предсказываем ltv пользователя (в деньгах), и если предсказанный ltv пользователя выше, чем затраты на этого пользователя (на его привлечение) больше, то почему бы не купить этого пользователя)

- Подготовка признаков для будущей модели

    - `feature extraction and selection` (создание и отбор признаков)

```python
contacts = pd.concat([
    pd.read_excel('contacts_with_man_gender.xlsx'), 
    pd.read_excel('contacts_with_na_gender.xlsx'), 
    pd.read_excel('contacts_with_woman_gender.xlsx'), 
    pd.read_excel('contacts_with_no_mobile.xlsx')]
)
```

Удалим из датасета с контактами агрегаторы и сотрудников. Информация по ним не отражает реальный портрет клиента и представляет собой, по большей части, выбросы. 

```python
staff_list = list(
    contacts[(contacts['atribute_values']=='Скидка 20%') | 
             (contacts['atribute_values']=='Скидка 25%') |
             (contacts['atribute_values']=='Скидка 10%') |
             (contacts['atribute_values']=='Скидка 30%') |
             (contacts['atribute_values']=='Скидка 50%') |
             (contacts['atribute_values']=='сотрудник')]['contact_id'].unique()
)
aggs = ('Яндекс', 'Яндекс Еда', 'Яндекс.Еда', 'Деливери')
agg_list = list(contacts[contacts['first_name'].isin(aggs)]['contact_id'].unique())
staff_list.extend(agg_list)
contacts = contacts.loc[~contacts['contact_id'].isin(staff_list)]
```

Удалим из датасета бессмысленные имена, которые фактически означают отсутствие имени, для того, чтобы лучше видеть качество данных в этой колонке, а также в дальнейшем для формирования признаков. 

```python
contacts.loc[contacts['first_name']=='Выдать анкету! Пустая карта', 'first_name'] = np.nan
contacts.loc[contacts['first_name']=='Без имени', 'first_name'] = np.nan
```

Превратим фамилию в `0` или `1`, т.к. она не имеет значения, а важным может быть то, захотел ли ее указать клиент или нет при заполнении анкеты (как знак лояльности) - как гипотеза. Так мы поступим и с отчеством, и с номером телефона и электронной почтой. Имя оставим, т.к. на его основе мы заполним часть пропусков пола.
